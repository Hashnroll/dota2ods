```python
import os
import pandas as pd
```

```python
bot_id='1c2500195e6ab646fc9371344724aa77'
```

```python
import json
PATH_TO_DATA = './mlcourse-dota2-win-prediction/'

winrate = {}

with open(os.path.join(PATH_TO_DATA, 'train_matches.jsonl')) as fin:
    
    for line in fin:
        match = json.loads(line)
        for slot, player in enumerate(match['players']):
            if slot < 5:
                radiant = True
            else:
                radiant = False
            id_hash = player['account_id_hash']
                    
            if not id_hash in winrate:
                winrate[id_hash] = {'games':1,'wins':0,'losses':0}
            else:
                winrate[id_hash]['games'] += 1
                
            if ((radiant) and (match['targets']['radiant_win'] == True))\
                or ((not radiant) and (match['targets']['radiant_win'] == False)):
                    winrate[id_hash]['wins']+=1
            else:
                winrate[id_hash]['losses']+=1
    
winrate
```

```python
winrate = {k: {'games':v['games'], 'winrate':v['wins']/v['games']} for k, v in winrate.items()}
```

```python
{k: v for k, v in winrate.items() if v['games']>10}
```

```python
test_players=set()
with open(os.path.join(PATH_TO_DATA, 'test_matches.jsonl')) as fin:
    
    for line in fin:
        match = json.loads(line)
        for player in match['players']:
            id_hash = player['account_id_hash']
                    
            test_players.add(id_hash)

test_players
```

```python
def read_matches(matches_file):
    
    MATCHES_COUNT = {
        'test_matches.jsonl': 10000,
        'train_matches.jsonl': 39675,
    }
    _, filename = os.path.split(matches_file)
    total_matches = MATCHES_COUNT.get(filename)
    
    with open(matches_file) as fin:
        for line in tqdm_notebook(fin, total=total_matches):
            yield json.loads(line)
```

```python
import collections

def extract_features_csv(match):
    row = [
        ('match_id_hash', match['match_id_hash']),
    ]
        
    for slot, player in enumerate(match['players']):
        if slot < 5:
            player_name = 'r%d' % (slot + 1)
        else:
            player_name = 'd%d' % (slot - 4)
            
        if player['account_id_hash'] == bot_id:
            row.append((f'{player_name}_bot', 1))
        else:
            row.append((f'{player_name}_bot', 0))
        
    return collections.OrderedDict(row)
```

```python
from tqdm import tqdm_notebook
import json
```

```python
%%time
bot_feature_train = []

for match in read_matches(os.path.join(PATH_TO_DATA, 'train_matches.jsonl')):
    match_id_hash = match['match_id_hash']
    features = extract_features_csv(match)
    
    bot_feature_train.append(features)
```

```python
bot_feature_train = pd.DataFrame.from_records(bot_feature_train).set_index('match_id_hash')
```

```python
%%time
bot_feature_test = []

for match in read_matches(os.path.join(PATH_TO_DATA, 'test_matches.jsonl')):
    match_id_hash = match['match_id_hash']
    features = extract_features_csv(match)
    
    bot_feature_test.append(features)
```

```python
bot_feature_test = pd.DataFrame.from_records(bot_feature_test).set_index('match_id_hash')
```

```python
bot_feature = pd.concat([bot_feature_train, bot_feature_test], sort=False)
```

```python
r_columns = [f'r{i}_bot' for i in range(1, 6)]
d_columns = [f'd{i}_bot' for i in range(1, 6)]

bot_feature['r_total_' + 'bot'] = bot_feature[r_columns].sum(1)
bot_feature['d_total_' + 'bot'] = bot_feature[d_columns].sum(1)
bot_feature['total_' + 'bot' + '_diff'] = bot_feature['r_total_' + 'bot'] - bot_feature['d_total_' + 'bot']

bot_feature['r_std_' + 'bot'] = bot_feature[r_columns].std(1)
bot_feature['d_std_' + 'bot'] = bot_feature[d_columns].std(1)
bot_feature['std_' + 'bot' + '_diff'] = bot_feature['r_std_' + 'bot'] - bot_feature['d_std_' + 'bot']

bot_feature['r_mean_' + 'bot'] = bot_feature[r_columns].mean(1)
bot_feature['d_mean_' + 'bot'] = bot_feature[d_columns].mean(1)
```

```python
bot_feature.drop(r_columns+d_columns, axis=1, inplace=True) 
```
